```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

ALVO_REAL = 50
NUM_AMOSTRAS = 5
CONFIANCA = 0.90

plt.axvline(x=ALVO_REAL, color='red', linestyle='--', linewidth=2, label=f'Média Real ({ALVO_REAL})') #Desenha linha vertical vermelha

for i in range(NUM_AMOSTRAS):
    dados = np.random.normal(ALVO_REAL, 10, 25) #Simula coleta de dados (25)

    ic = st.t.interval(CONFIANCA, df=24, loc=np.mean(dados), scale=np.std(dados, ddof=1)/np.sqrt(25))

    lower, upper = ic

    cor = 'green' if lower <= ALVO_REAL <= upper else 'blue'

    plt.plot([lower, upper], [i, i], color=cor, linewidth=4, marker='|', markersize=10) #Estimativa Intervalar
    plt.plot(np.mean(dados), i, 'o', color='black', markersize=5) #Estimativa Pontual

plt.title(f'A Rede de Captura: {NUM_AMOSTRAS} Intervalos de Confiança ({CONFIANCA*100:.0f}%)')
plt.xlabel('Valor Estimado')
plt.legend()
plt.show()
```

```python
#TEOREMA CENTRAL DO LIMITE (TCL) - ESTIMATIVA PONTUAL
import numpy as np
import matplotlib.pyplot as plt

# Simula 1000 amostras (tamanho 30) de uma população
# com média 50 e desvio padrão 10.
means = [np.mean(np.random.normal(50, 10, 30)) for _ in range(1000)]

plt.hist(means, bins=30)
plt.title('Distribuição das estimativas pontuais (média)')
plt.xlabel('Média Amostral')
plt.ylabel('Frequência')
plt.show()

##O gráfico representa a Distribuição das Médias Amostrais (estimativas pontuais) tiradas de uma população, demonstrando que a maioria dessas estimativas se agrupa em torno da média populacional real.
```
