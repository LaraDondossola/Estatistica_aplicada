```python
import numpy as np

matriz = np.array([[1,2,3],[4,5,6]])

resultado = matriz * 2

print("Normal:\n", matriz)
print("\nMultiplicado por 2:\n", resultado)
```

```python
import pandas as pd
from pandas import Series
from pandas import DataFrame

obj = Series([4,7,-5,3])

obj #fomando a Series visualmente (em matriz)

type(obj) #para ver o tipo de objeto que está usando

obj.values #para verificar os valores que estão dentro do objeto

obj.index #mostra os índices, começo, fim, percorre os índices de 1 em 1

```

```python
obj2 = Series([4,7,-5,3], index = ['a', 'b', 'c', 'd'])

obj2

obj2.values

obj2.index

obj2[obj2>3] #somente maior que 3

obj2['b'] #vai buscar por posição específica

'd' in obj2 #perguntar se a posição 'd' existe
```

```python
dict = {'futebol':5200, 'tenis':120, 'natação':698,  'volleyball':150}

obj3 = Series(dict)

obj3
```

```python
esportes = ['futebol', 'tenis', 'natação','basketball']

obj4 = Series(dict, index = esportes)

obj4

pd.isnull(obj4)

pd.notnull(obj4)
```

```python
obj3 + obj4 #soma pelos índices
```

```python
obj4.name = 'população'
obj4.index.name = 'esporte'

obj4
```

```python
from pandas import DataFrame

data = {'estado': ['Santa Catarina', 'Paraná', 'Goiás', 'Bahia', 'Minas Gerais'],
        'ano': [2002,2003,2004,2005,2006],
        'população':[1.5,1.7,3.6,2.4,2.9]}

frame = DataFrame(data)

type(frame)

frame
```

```python
#para mudanças no formato, mudar as colunas, PROVA

DataFrame(data, columns= ['ano', 'estado', 'população'])
```

```python
#para criar um frame maior

frame2 = DataFrame(data, columns= ['ano', 'estado','população', 'débito'], index = ['um', 'dois', 'três', 'quatro', 'cinco'])

frame2
```
