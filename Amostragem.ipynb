```python
def amostragem_sistematica(dataset, amostras): #crio uma função que posso utilizar a qualquer momento pra me retornar amostragens.
  intervalo=len(dataset)//amostras  #vai descobrir qual o intervalo ideal, dividindo a populacao pra saber a cada quantos vou retirar 1 elemento
  random.seed(1) #fixo seed para conseguir repetir sempre igual (dentro da mesma populacao)
  inicio=random.randint(0, intervalo) #pego aleatoriamento 1 item dentro do primeiro intervalo
  indices = np.arange(inicio, len(dataset), intervalo) #pego o restante dos elementos, indo até o final da populacao, pulando de acordo com intervalo encontrado
  amostra_sistematica=dataset.iloc[indices] #coloco o vetor feito dentro da variável
  return amostra_sistematica #retorno esse vetor para quem estiver chamando a função
```

```python
df_amostra_sistematica=amostragem_sistematica(dataset, 100) #chama a funcao que criamos, passando que quer 100 elementos
df_amostra_sistematica.shape #mostra que realmente retorno 100 elementos
```
